```r

View(Zadan10)

attach(Zadan10)

names(Zadan10)

diziz_imp = as.data.frame(scale(na.omit(Zadan10)))

diziz.pcs = princomp(~x1+x2+x3+x4+x5+z1+z2+t1+t2+t3+t4, data= diziz_imp)
summary(diziz.pcs)
#из результатов видим что 1 компанента объеснят 64% общей вариации данных 
#а первые 4 - 94%


#оценка необходимого числа главных компонет 
ev = diziz.pcs$sdev
ev
ev[ev >  mean(ev)]# криейрий Кайзера-Гутманта 
#оставим для дальнейшего анализа тольео те значения кторые 
#превышают (стеднее собственных значений)

#проилюстрируем это на графике, где линия это среднее значение 
barplot(ev)
abline(h = mean(ev), col = 'red')
# Анализуруя таблицу можно предположить что 2 компанента связанна 
# с переменными (z2,t2); а 3 с (z1, x5); 4(t1);

P = loadings(diziz.pcs)[,1:4]
P
#       Comp.1     Comp.2       Comp.3       Comp.4
# x1  0.3575912  0.1678637  0.008784587  0.144264730
# x2  0.3613642  0.1321654 -0.040376451  0.235582582
# x3 -0.3579336  0.1713425 -0.026492447 -0.229588755
# x4 -0.3507593  0.1675978 -0.182008559 -0.233746911
# x5 -0.2701548  0.1307049  0.675032406! -0.134268552
# z1  0.3547712  0.1543069 -0.081910087 -0.008705201
# z2 -0.0102386  0.5793315! -0.516355574! -0.314672939
# t1  0.0752731 -0.5823550! -0.165796370 -0.588548761!
# t2 -0.2971401  0.2883939  0.126635246  0.197115956
# t3 -0.2965246 -0.2828887 -0.310620180  0.492737758
# t4 -0.3309982 -0.1343612 -0.308954202  0.266799914

#составим матрицу счетов Т
T= predict(diziz.pcs)[,1:4]
T

diziz_PCA = as.data.frame(cbind(y = diziz_imp[,6],T))
#diziz_PCA 
#diziz_imp[['y']]
M.PCA = lm(y~., data = diziz_PCA  )
summary(M.PCA)
#по результатам выделим значимые компоненты
#  Comp.1    Pr(>F) = 2.06e-12 <0.05
#  Comp.2    Pr(>F) = 0.000842 <0.05
#p-value: 6.115e-11 <0.05 

AIC(M.PCA)
#10.3986
#Полученная модель регресии на главных компонент - удачная по величине AIC

rss.M.PCA = sum(M.PCA$resid^2)/M.PCA$df.residual
rss.M.PCA
#0.06978191 - оценка модели но RSS оценка качества не всегла оправдано.


# 
# library(pls)
# 
# M.pls = plsr(y~., 5, data = diziz_imp, method = 'oscorespls')
# summary(M.pls)
# 
# 
# beta.pls = drop(coef(M.pls))
# beta.pls
# resid.pls = drop(M.pls$resid)[,5]
# rss.pls = sum(resid.pls^2)/(nrow(diziz_imp)- 5)
# 
# rss.pls
```
