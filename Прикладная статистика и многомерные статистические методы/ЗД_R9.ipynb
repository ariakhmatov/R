```r
#1. Анилиз влияния грунта и цвета
table(priming, view)
#по таблице видим одинаковые значения на всех эксперементальных групах
#это означает о сбалансированности

aggregate(weight, by = list(priming, view), FUN = mean)
#вычислим среднее отклонение для каждой группы 

aggregate(weight, by = list(priming, view), FUN = sd)
#вычислим стандартные отклонение для каждой группы 

fit = aov(weight ~ priming*view) #дисперсионный анализ 
summary(fit)#проверим значимость взаимодействия факторов

#Pr(>F) для грунта = 0.0496 < 0,05 значимо 
#Pr(>F) для цвета = 0.2779 > 0,05   не значимо 
#Pr(>F) для грунта:цвета = 0.2713 > 0,05 не значимо 

interaction.plot(priming, view, weight, type='b', col=c('red','blue'))
#На диаграмме видно средний вес
#при всех трех грунтах и двух его цвета.

boxplot(weight ~ priming*view)
# на этом графике видно среднее значение веса в соочетаний двух факторов

#Вывод влияния грунта и цвета на Вес Иглокожих.
#фактор грунта влияем на вес, а цвет нет.
```

```r
#2. Анилиз влияния Грунта и Района
table(priming, region)
#по таблице видим одинаковые значения на всех эксперементальных групах
#это означает о сбалансированности

aggregate(weight, by = list(priming, region), FUN = mean)
#вычислим среднее отклонение для каждой группы 

aggregate(weight, by = list(priming, region), FUN = sd)
#вычислим стандартные отклонение для каждой группы 

fit = aov(weight ~ priming*region) #дисперсионный анализ 
summary(fit)#проверим значимость взаимодействия факторов

#Pr(>F) для грунта =  0.0422 < 0,05 значимо 
#Pr(>F) для региона = 0.1226 > 0,05   не значимо 
#Pr(>F) для грунта:региона = 0.3257 > 0,05 не значимо 

interaction.plot(priming, region, weight, type='b', col=c('red','blue'))
#На диаграмме видно средняя вес
#при всех трех грунтках и трех районах.(в popov выше вес в среднем)

boxplot(weight ~ priming*region)
# на этом графике видно среднее значение веса в соочетаний двух факторов

#Вывод влияния грунта и района на Вес Иглокожих.
#фактор грунта влияет на вес, а район нет.
```

```r
#3. Анилиз влияния Грунта и Условия
table(priming, conditions)
#по таблице видим одинаковые значения на всех эксперементальных групах
#это означает о сбалансированности

aggregate(weight, by = list(priming, conditions), FUN = mean)
#вычислим среднее отклонение для каждой группы 

aggregate(weight, by = list(priming, conditions), FUN = sd)
#вычислим стандартные отклонение для каждой группы 

fit = aov(weight ~ priming*conditions) #дисперсионный анализ 
summary(fit)#проверим значимость взаимодействия факторов

#Pr(>F) для грунта =  0.0543 < 0,05 не значимо 
#Pr(>F) для условия = 0.1240 > 0,05   не значимо 
#Pr(>F) для грунта:условие= 0.8595 > 0,05 не значимо 

interaction.plot(priming, conditions, weight, type='b', col=c('red','blue'))
#На диаграмме видно средняя вес
#при всех трех грунтах и двух условий.(в срденем natural выше вес)

boxplot(weight ~ priming*conditions)
# на этом графике видно среднее значение веса в соочетаний двух факторов

#Вывод влияния грунта и условия на Вес Иглокожих.
#фактор грунта, условия не влияют на вес.
```

```r
#2.Многомерный дисперсиозный анализ 

#Зависимость выбора университета от рейтингов
y = cbind(Yslov_educ,Yslov_job, Yslov_science)

#рейтин университета пусть будет назависеммой перемменной(Reyt_funcz)
#тогда получим групировку вузов по общему рейтингу
aggregate(y, by =  list(Reyt_funcz), FUN = mean)
```

```r
cov(y)# оценки ковариациооной матрицы 
# по разультатам видно что отношение разных рангов с друг другом
```

```r
fit1 = manova(y ~ Reyt_funcz)#многомерный тест на межгрупповые различия
summary(fit1)
#Pr(>F) = 3.6e-05 < 0.05 значимы 
```
